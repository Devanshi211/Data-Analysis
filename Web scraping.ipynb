```python
import requests
url = 'https://www.allsides.com/media-bias/media-bias-ratings'
r=requests.get(url)
print(r.content[:100])
```

```python
!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.content,'html.parser')
```

```python
rows = soup.select('tbody tr')
print(rows)
```

```python
print(rows[0])
```

```python
row1 = rows[0]
name = row1.select_one('.views-field-title').text.strip()
print(name)
```

```python
allsides_page = row1.select_one('.source-title a')['href']
allsides_page = 'https://www.allsides.com' + allsides_page
print(allsides_page)
```

```python
bias = row1.select_one('.views-field-field-bias-image a img')['title']
bias = bias.split(':')
print(bias[-1].strip())
```

```python
agree = row1.select_one('.agree').text.strip()
print("People agreed:"+agree)
disagree = row1.select_one('.disagree').text.strip()
print("People disagreed:"+disagree)
ratio = int(agree)/int(disagree)
print("%.2f" %ratio)
```

```python
feedback = row1.select_one('.hidden-xs')
print(feedback)
```

```python
def getrating(ratio):
    if(ratio>3):
        return "Absolutely agrees"
    elif(ratio>2 and ratio<=3):
        return "Strongly agrees"
    elif(ratio>1.5 and ratio<=2):
        return "agrees"
    elif(ratio>1 and ratio<=1.5):
        return "somewhat agrees"
    elif(ratio>0.67 and ratio<1):
        return "Strongly agrees"
    elif(ratio>0.5 and ratio<=0.67):
        return "Strongly agrees"
    elif(ratio>0.33 and ratio<=0.5):
        return "Strongly agrees"
    elif(ratio<=0.333):
        return "Strongly agrees"
    else:
        return "Neutral"
    
    
print(getrating(5))
```

```python
data = []
for row in rows:
    d = dict()
    
    d['name'] = row.select_one('.views-field-title').text.strip()
    d['allsides_page'] = 'https://www.allsides.com' + row.select_one('.source-title a')['href']
    d['bias'] = row.select_one('.views-field-field-bias-image a img')['title'].split(':')[-1]
    d['agree'] = int(row.select_one('.agree').text)
    d['disagree'] = int(row.select_one('.disagree').text)
    d['ratio'] = d['agree'] / d['disagree']
    d['rating'] = getrating(d['ratio'])
    
    data.append(d)

print(data[0])
```
